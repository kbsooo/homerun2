```python
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_csv('gStation_standard.csv')
```

```python
# 가상의 날짜 부여
dates = pd.date_range(start='2025-03-03', periods=len(df), freq='D')
df['ds'] = dates
```

```python
# 출발 시간을 숫자로 변환 (SCHOOL_DEPART COL 예측한다고 가정)
def time_to_float(time_str):
    hours, minutes = map(int, time_str.split(':'))
    return hours + minutes / 60.0

df['y'] = df['MON_SCHOOL_DEPART'].apply(time_to_float)
```
